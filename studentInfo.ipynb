```python
# Melihat Dataset secara garis besar
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/studentInfo.csv')
df
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Mengisi Null Values yang ada di kolom 'imd_band' dengan nilai 0
df['imd_band'].fillna(0, inplace=True)
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Visualisasi Bar Chart banyak data dari final_result
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.histplot(df['final_result'], bins=15)
plt.xlabel('final_result')
plt.ylabel('Count')
plt.title('Count of final_result')
plt.show()
```

```python
# Visualisasi Bar Chart banyak data dari 'gender'
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(x='gender', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Gender')
plt.xticks(rotation=45)
plt.show()
```
