```python
# Melihat Dataset secara garis besar
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/courses.csv')
df
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Visualisasi Box Plot untuk melihat data outlier dari 'module_presentation_length'
import matplotlib.pyplot as plt

df['module_presentation_length'].plot(kind="box", figsize = (6,4))
plt.title('Box Plot of Module Presentation Length\n', size=16)
plt.show()
```

```python
# Visualisasi Box Plot hubungan antara 'code_module' dengan 'module_presentation_length'
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='code_module', y='module_presentation_length', data=df)
plt.title('Hubungan Code Module dan Module Presentation Length')
plt.xlabel('Code Module')
plt.ylabel('Module Presentation Length')
plt.show()
```

```python
# Visualisasi Box Plot hubungan antara 'code_presentation' dengan 'module_presentation_length'
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='code_presentation', y='module_presentation_length', data=df)
plt.title('Hubungan Code Presentation dan Module Presentation Length')
plt.xlabel('Code Module')
plt.ylabel('Module Presentation Length')
plt.show()
```

```python
# Visualisasi Bar Chart banyak data dari code_presentation
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=df)
plt.xlabel('Code Presentation')
plt.ylabel('Count')
plt.title('Count of Data per Code Presentation')
plt.xticks(rotation=45)
plt.show()
```
