```python
# Melihat Dataset secara garis besar
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/vle.csv')
df
```

```python
# Menghapus kolom yang tidak perlu karena null values yaitu 'week_form' dan 'week_to'
df = df.drop('week_from', axis=1)
df = df.drop('week_to', axis=1)
```

```python
df
```

```python
# Visualisasi Bar Chart banyak data dari 'activity_type'
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(x='activity_type', data=df)
plt.xlabel('activity_type')
plt.ylabel('Count')
plt.title('Count of activity_type')
plt.xticks(rotation=45)
plt.show()
```
