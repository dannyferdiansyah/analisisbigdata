```python
# Melihat Dataset secara garis besar
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/studentVle.csv')
df
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Visualisasi data menggunakan heatmap untuk memvisualisasikan distribusi sum_click berdasarkan kombinasi code_module
import matplotlib.pyplot as plt
import seaborn as sns

pivot_table = df.pivot_table(values="sum_click", index="code_module", columns="code_presentation")
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("Distribusi sum_click per code_module dan code_presentation")
plt.xlabel("code_presentation")
plt.ylabel("code_module")
plt.show()
```
