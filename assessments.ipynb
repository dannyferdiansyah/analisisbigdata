```python
# Melihat Dataset secara garis besar

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/assessments.csv')
df
```

```python
# Melihat/showing percentage null values yang ada di dataset

print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Melakukan interpolasi linier pada kolom tertentu (Null Values)
df['date'] = df['date'].interpolate(method='linear')
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Visualisasi Box Plot untuk melihat data outlier dari 'weight'
import matplotlib.pyplot as plt

df['weight'].plot(kind="box", figsize = (6,4))
plt.title('Box Plot of Assessments Weight\n', size=16)
plt.show()
```

```python
# Visualisasi Bar Chart banyak data dari code_module
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='code_module', data=df)
plt.xlabel('Code Module')
plt.ylabel('Count')
plt.title('Count of Data per Code Module')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi Bar Chart banyak data dari code_presentation
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=df)
plt.xlabel('Code Presentation')
plt.ylabel('Count')
plt.title('Count of Data per Code Presentation')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi Box Plot hubungan antara 'code_module' dengan 'weight'
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='code_module', y='weight', data=df)
plt.title('Hubungan Code Module dan Weight')
plt.xlabel('Code Module')
plt.ylabel('Weight')
plt.show()
```

```python
# Visualisasi Box Plot hubungan antara 'code_presentation' dengan 'weight'
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='code_presentation', y='weight', data=df)
plt.title('Hubungan Code Presentation dan Weight')
plt.xlabel('Code Presentation')
plt.ylabel('Weight')
plt.show()
```

```python
# Visualisasi Box Plot hubungan antara 'assessment_type' dengan 'weight'
plt.figure(figsize=(12, 6))
sns.boxplot(x='assessment_type', y='weight', data=df)
plt.title('Hubungan Assessment Type dan Weight')
plt.xlabel('Assessment Type')
plt.ylabel('Weight')
plt.show()
```
