```python
# Melihat Dataset secara garis besar
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BIG-DATA/studentRegistration.csv')
df
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Mengisi Null Values yang ada di kolom 'date_unregistration' dan 'date_registration' dengan nilai "Uncategorized"
df["date_unregistration"].fillna("Uncategorized", inplace = True)
df["date_registration"].fillna("Uncategorized", inplace = True)
```

```python
# Melihat/showing percentage null values yang ada di dataset
print(f"Data Types:\n{df.dtypes}")
print(f"\nRows and Columns:\n{df.shape}")
print(f"\nColumn Names:\n{df.columns}")
print(f"\nNull Values %:\n{df.apply(lambda x: sum(x.isnull()) / len(df))}")
```

```python
# Visualisasi Bar Chart banyak data dari 'code_module'
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(x='code_module', data=df)
plt.xlabel('code_module')
plt.ylabel('Count')
plt.title('Count of code_module')
plt.xticks(rotation=45)
plt.show()
```
